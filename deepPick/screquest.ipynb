```python
from obspy import UTCDateTime
from obspy.clients.fdsn import Client
from obspy.core.event import ResourceIdentifier
import numpy as np

server_IP = 'http://10.100.100.232'
port_fdsn = "8091"
    

starttime = UTCDateTime("2016 09 01 000000")    

endtime = UTCDateTime("2016 10 01 000000")

client = Client(server_IP+":"+port_fdsn)
```

```python
# catálogo de eventos correspondiente al select
cat = client.get_events(starttime=starttime, endtime=endtime, minmagnitude=5)
print len(cat)
```

```python
# seleccionando el primer evento del catálogo
event = cat[1]
print event
```

```python
# seleccionando el origen preferido del evento
pref_origin = event.preferred_origin()
print pref_origin
```

```python
print event.preferred_origin_id.id

event.preferred_origin_id = ResourceIdentifier(id=event.preferred_origin_id.id.replace("#","_").replace("@","a"))
event.preferred_magnitude_id = ResourceIdentifier(id=event.preferred_origin_id.id.replace("#","_").replace("@","a"))
print event.preferred_origin_id.id

event.write("sismo2.6.xml", format="QUAKEML", validate=True)
```
